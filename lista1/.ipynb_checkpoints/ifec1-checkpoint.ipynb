```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Questão (a)

**Funções**

```python
def random_walk_1D(n):
  dim = 1
  x = 0
  y = 0
  xposition = [0]
  yposition = [0]

  values_range = [-1/2, 1/2]

  # gerando valores aleatórios entre -1/2 e 1/2, em uma dimensão
  steps = np.random.uniform(values_range[0], values_range[1], (n, dim))

  # somando os valores gerados em y enquanto somo 1 ao eixo x afim de obtermos uma boa vizualização
  for i in range (0, n):
    x += 1
    y += steps[i][0]
 
    xposition.append(x)
    yposition.append(y)

  return [xposition,yposition]

def random_walk_2D(n):
  dim = 2
  x = 0
  y = 0
  xposition = [0]
  yposition = [0]

  values_range = [-1/2, 1/2]

  # gerando valores aleatórios entre -1/2 e 1/2, em duas dimensões
  steps = np.random.uniform(values_range[0], values_range[1], (n, dim))

  # somando os valores gerados tanto em y quanto em x
  for i in range(0, n):
    # pegando numero aleatorio (0 ou 1) afim de decidir em qual eixo meu passo ocorrerá
    randomize = np.random.randint(0, 2) 

    if randomize == 0:
      x += steps[i][0]
    else: 
      y += steps[i][1]

    xposition.append(x)
    yposition.append(y)

  return [xposition,yposition]



def random_walk_many_times():
  dim = 1
  y = 0
  x = 0
  values_range = [-1/2, 1/2]

  # gerando valores aleatórios entre -1/2 e 1/2, em uma dimensão para 100 passos
  steps = np.random.uniform(values_range[0], values_range[1], (100, dim))

  for i in range (0, 100):
    y += steps[i][0]



  # gerando valores aleatórios entre -1/2 e 1/2, em uma dimensão para 1 passo
  steps = np.random.uniform(values_range[0], values_range[1], (1, dim))
  
  for i in range (0, 1):
    x += steps[i][0]


  return [y, x]
```

**Plotando os gráficos**

**1 dimensão**

```python
random_walk_1 = random_walk_1D(10)
plt.plot(random_walk_1[0],random_walk_1[1])
plt.title("Random Walk 1-D - 10")
plt.show()

random_walk_1 = random_walk_1D(1000)
plt.plot(random_walk_1[0],random_walk_1[1])
plt.title("Random Walk 1-D - 1000")
plt.show()

random_walk_1 = random_walk_1D(100000)
plt.plot(random_walk_1[0],random_walk_1[1])
plt.title("Random Walk 1-D - 100000")
plt.show()
```

**2 dimensão**

```python
random_walk_2 = random_walk_2D(10)
plt.plot(random_walk_2[0], random_walk_2[1])
plt.title("Random Walk 2-D - 10")
plt.show()

random_walk_2 = random_walk_2D(1000)
plt.plot(random_walk_2[0], random_walk_2[1])
plt.title("Random Walk 2-D - 1000")
plt.show()

random_walk_2 = random_walk_2D(100000)
plt.plot(random_walk_2[0], random_walk_2[1])
plt.title("Random Walk 2-D - 100000")
plt.show()
```

**Se você multiplicar o número de passo por 100, a distância final da caminhada aumenta por cerca de 10 vezes?**


```python
media = 0
for i in range(0, 10000000+1):
  walks = random_walk_many_times()
  razao = walks[0]/walks[1]
  media += razao
  # print("Teste", i, "- razão:", razao)
  if i == 10:
    print("Razão média para 10 tentativas:", abs(media/10))
  elif i == 100:
    print("Razão média para 100 tentativas:", abs(media/100))
  elif i == 1000:
    print("Razão média para 1000 tentativas:", abs(media/1000))
  elif i == 10000:
    print("Razão média para 10000 tentativas:", abs(media/10000))
  elif i == 100000:
    print("Razão média para 100000 tentativas:", abs(media/100000))
  elif i == 1000000:
    print("Razão média para 1000000 tentativas:", abs(media/1000000))
  elif i == 10000000:
    print("Razão média para 10000000 tentativas:", abs(media/10000000))

```

Em média, a caminhada é a raiz do número de passos. Como forma de prova, optei pela prática, realizando 10000000 testes em 1D com caminhadas de 100 e de 1 passos, como podemos ver acima, comparando a razão entre as caminhadas em seu ponto final.


# Questão (b)

```python
def random_walk_2D_with2steps(n):
  dim = 2
  x = 0
  y = 0
  xposition = [0]
  yposition = [0]

  values_range = [-1/2, 1/2]

  # gerando valores aleatórios entre -1/2 e 1/2, em duas dimensões
  steps = np.random.uniform(values_range[0], values_range[1], (n, dim))

  # somando os valores gerados tanto em y quanto em x
  for i in range(0, n):
    # diferentemente da função da questão anterior, aqui, puramente para questão de teste, modifico tanto x quanto y por passo, pois caso contrário o teste com 1 passo seria sempre um dos eixos no zero ex.: [0, 0.2] / [-0.25, 0]
    x += steps[i][0]
    y += steps[i][1]

    xposition.append(x)
    yposition.append(y)

  return [xposition,yposition]
```

```python
xposition_one_step = []
yposition_one_step = []

xposition_ten_steps = []
yposition_ten_steps = []

for i in range (0, 10000):
  one_step = random_walk_2D_with2steps(1)
  ten_steps = random_walk_2D_with2steps(10)

  # adiciono em meus arrays principais os xs e ys gerados para cada quantidade de passos por meio do extend
  xposition_one_step.extend(one_step[0])
  yposition_one_step.extend(one_step[1])

  xposition_ten_steps.extend(ten_steps[0])
  yposition_ten_steps.extend(ten_steps[1])

plt.scatter(xposition_ten_steps, yposition_ten_steps, s=0.5, c = 'blue', alpha=0.5)
plt.scatter(xposition_one_step, yposition_one_step, s=0.5, c = 'black', alpha=0.5, marker = 'o')
```

# Questão (c)

```python
def variance(arr):
  mean = arr.mean()
  soma = 0
  for i in range(len(arr)):
    soma += (arr[i]-mean)*(arr[i]-mean)

  return soma/(len(arr)-1)


def rms(w):
  dim = 1
  values_range = [-1/2, 1/2]
  # gerando valores aleatórios entre -1/2 e 1/2, em uma dimensão
  steps = np.random.uniform(values_range[0], values_range[1], (w, dim))
  return math.sqrt(variance(steps))

def generateHist():
  w = 10000
  n = [1, 2, 3, 5]
  last_ones = []

  for i in n:
    for q in range(0, w):
      positions = random_walk_1D(i)
      last_ones.append(positions[1][-1])

    rms_value = rms(w)
    sigma = math.sqrt(i)*rms_value
    interval = (-3*sigma, 3*sigma)
    x = np.arange(start=float(interval[0]), stop=float(interval[1]), step=float(interval[1]/w))
    y = []

    for j in range(0, len(x)):
      inferior_exp = 2*(sigma**2)
      exp = np.exp(-(x[j]**2)/inferior_exp)
      inferior = math.sqrt(2*np.pi)*sigma
      y.append(exp/inferior)


    plt.hist(last_ones, bins=50, density=True, edgecolor='black', zorder=1)
    plt.title('N = ' + str(i) + ' steps')
    plt.xlabel("Pontos finais")
    plt.scatter(x, y, s=0.5 , c="red", alpha=0.5, marker='o',zorder=2)
    plt.show()




```

```python


generateHist()
```

Nessa etapa não pude conseguir algo tão próximo de uma Gaussiana. Em meus testes, minhas caminhadas tendem ao centro e destoam do esperado. Pude perceber isso aumentando os valores de n e de w.
