```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
def walk1D(steps_number):
    step_size = [1/2, -1/2]
    origin = np.zeros((1,1))
    dims = [steps_number, 1]
    steps_vector = np.random.choice(step_size, dims)
    path = np.concatenate([origin, steps_vector]).cumsum(0)
    
    fig = plt.figure(figsize = (10,5))
    plt.title('1D random walk')
    ax = fig.add_subplot(111)
    ax.plot(path)
```

O número de passos é representado pelo eixo X do gráfico, enquanto a distância em relação à origem é representada pelo eixo Y. Dessa forma fica mais fácil de ver a mudança da distância ao longo da caminhada ao invés de termos apenas uma linha uni-dimensional

```python
walk1D(10000)
```

```python
def walk2D(steps_number):
    step_size = [1/2, 0, -1/2]
    origin = np.zeros((1,2))
    dims = [steps_number, 2]
    steps_vector = np.random.choice(step_size, dims)
    path = np.concatenate([origin, steps_vector]).cumsum(0)
    final = path[-1:]
    
    fig = plt.figure(figsize = (10,10))
    plt.title('2D random walk ' + str(steps_number) + ' steps')
    ax = fig.add_subplot(111)
    ax.plot(path[:,0], path[:,1],c='blue', alpha=0.5);
    ax.plot(final[:,0], final[:,1],c='black', marker='x')
    plt.gca().set_aspect('equal')
```

```python
walk2D(10)
```

```python
walk2D(100)
```

```python
walk2D(10000)
```

Podemos ver que nos exemplos gerados, a distância do ponto final (marcado com o X) em relação à origem não aumentou na mesma taxa que o número de passos, mas quando multiplicamos o número de passos por 100, a distância aumentou cerca de 10 vezes.

```python
def final(N, W):
    finals = np.zeros((W, 2))
    for i in range (1, W):
        step_size = [1/2, 0, -1/2]
        origin = np.zeros((1,2))
        dims = [N, 2]
        steps_vector = np.random.choice(step_size, dims)
        path = np.concatenate([origin, steps_vector]).cumsum(0)
        final = path[-1:]
        finals[i:] = final
    return finals
```

```python
ten_steps = final(10, 10000)
one_step = final(1, 10000)

fig = plt.figure(figsize = (10,10))
plt.title('Random walks final points dispersion')
##plt.title('2D random walk ' + str(N) + ' steps')
ax = fig.add_subplot(111)
ax.scatter(ten_steps[:,0], ten_steps[:,1],c='blue',alpha=0.25,s=0.5);
bx = fig.add_subplot(111)
bx.scatter(one_step[:,0], one_step[:,1], c='orange')

```

O centro do quadrado também está marcado por causa do zero nas possíveis escolhas para a caminhada de um passo. Ele foi incluído como artifício para possibilitar caminhadas na diagonal, mas acabou causando essa interação com a caminhada de um passo só. Isso não tem efeito nas caminhadas em si, apenas "diminui" o total de caminhadas feitas.

```python
def RMS_hist(N, W):
    finals = np.zeros((W, 1))
    for i in range (1, W):
        step_size = [1/2, 0, -1/2]
        origin = np.zeros((1,1))
        dims = [N, 1]
        steps_vector = np.random.choice(step_size, dims)
        path = np.concatenate([origin, steps_vector]).cumsum(0)
        final = path[-1]
        finals[i] = final
    square = np.square(finals)
    mean_square = np.mean(square)
    rms = np.sqrt(mean_square)
    sigma = np.sqrt(N) * rms
    mu = 0
    x = np.linspace(mu - 3 * sigma, mu + 3 * sigma)
    plt.hist(finals, density = True, align = 'mid', bins = 50)
    plt.plot(x, stats.norm.pdf(x, mu, sigma))
    plt.show()
```

```python
RMS_hist(1, 10000)
```

```python
RMS_hist(2, 10000)
```

```python
RMS_hist(3, 10000)
```

```python
RMS_hist(5, 10000)
```
